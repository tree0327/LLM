# AI 에이전트

## AI 에이전트란 무엇인가?


AI 에이전트는 **환경과 상호 작용하고, 데이터를 수집하고, 데이터를 사용하여 사전 결정된 목표를 달성하기 위해 필요한 작업을 스스로 결정해서 수행할 수 있는 소프트웨어 프로그램**이다.

* **자율성**: 사람이 목표를 설정하면 AI 에이전트는 목표를 달성하기 위해 필요한 최적의 조치를 독립적으로 선택한다.
* **지능적 판단**: 상황에 따라 고객 문의 자체를 해결할지, 사람에게 전달할지 결정하는 능력을 갖춘다.

### 예시: AI 상담원

* 자동으로 고객에게 여러 질문을 함
* 내부 문서의 정보를 조회함
* 해결책을 찾아 대응함
* 고객의 대답에 따라 적절한 다음 단계를 결정함

## AI 에이전트 동작 시나리오

### 실제 활용 예시

**사용자 요청**: "200만원의 예산내에서 10월 15일 출국 10월 24일 귀국 일정으로 도쿄로 가는 항공편과 숙소, 렌트카를 예약해줘."

### AI 에이전트가 활용하는 도구들

* **항공권 검색**: Skyscanner
* **숙소 예약**: Booking.com, Agoda
* **렌트카**: Rentalcars.com
* **일정 관리**: Google Calendar
* **정보 검색**: 네이버 항공권

AI 에이전트는 이러한 다양한 도구들을 **자동으로 연동**하여 사용자의 복잡한 요청을 처리한다.

## AI 에이전트의 장점

### 1. 생산성 향상

* 사람의 개입 없이 특정 작업을 수행하는 자율 지능형 시스템이다.
* 반복 작업을 AI 에이전트에게 위임하여 생산성을 향상시킨다.
* 중요한 활동이나 창의적인 활동에 집중할 수 있도록 지원한다.

### 2. 비용 절감

* 프로세스 비효율성, 사람의 오류, 수동 프로세스로 인한 불필요한 비용을 감소시킨다.
* 변화하는 환경에 적응하는 일관된 모델로 복잡한 작업을 안정적으로 수행한다.

### 3. 정보에 입각한 의사 결정

* 기계 학습을 활용하여 대량의 실시간 데이터를 수집하고 처리한다.
* 빠른 속도로 보다 정확한 예측이 가능하다.
* 다양한 시장 세그먼트의 제품 수요 분석 등 전략적 의사결정을 지원한다.

### 4. 고객 경험 개선

* 추천 제품을 개인화한다.
* 즉시 대응 시스템을 제공한다.
* 혁신을 통한 고객 참여, 전환율, 충성도를 향상시킨다.

## AI 에이전트의 작동과정

### 1. 목표 결정

* 사용자로부터 구체적인 명령 또는 목표를 받는다.
* 목표를 여러 개의 실행 가능한 작은 작업으로 분해한다.
* 특정 순서 또는 조건에 따라 작업 수행 계획을 수립한다.

### 2. 정보 획득

* 계획한 작업을 성공적으로 수행하기 위한 정보를 수집한다.
* 인터넷 검색을 통해 필요한 정보를 검색한다.
* 다른 에이전트 또는 기계 학습 모델과 상호 작용하여 정보를 교환한다.

### 3. 작업 구현

* 충분한 데이터가 있으면 당면한 작업을 체계적으로 구현한다.
* 작업 완료 후 목록에서 해당 작업을 제거하고 다음 작업을 진행한다.
* 외부 피드백을 받고 자체 로그를 검사하여 목표 달성 여부를 평가한다.
* 필요시 추가 작업을 생성하고 실행한다.

## AI 에이전트의 구성요소

### 핵심 구성 요소

* **Agent (에이전트)**: 중앙 제어 시스템
* **Memory (메모리)**: 단기 기억과 장기 기억
* **Tools (도구)**: Calendar, Calculator, CodeInterpreter, Search 등
* **Planning (계획)**: 작업 계획 수립
* **Action (행동)**: 실제 작업 수행

### 고급 기능

* **Reflection (반성)**: 수행 결과 검토
* **Self-critics (자기 비판)**: 성능 개선을 위한 자체 평가
* **Chain of thoughts (사고 연쇄)**: 논리적 추론 과정
* **Subgoal decomposition (하위 목표 분해)**: 복잡한 목표를 작은 단위로 분해

## AI 에이전트의 어려운점

### 1. 데이터 프라이버시 문제

* 대량의 데이터 획득, 저장, 이동이 필요하다.
* 데이터 프라이버시 요구 사항 충족이 필요하다.
* 데이터 보안 태세 강화 조치가 필요하다.

### 2. 윤리적 과제

* 딥 러닝 모델의 불공정하거나 편향된 결과가 발생할 가능성이 있다.
* 사람의 검토와 같은 보호 조치가 필요하다.
* 공정한 응답을 위한 추가적인 안전장치 구현이 필요하다.

### 3. 기술적 복잡성

* 기계 학습 기술에 대한 전문적인 경험과 지식이 필요하다.
* 기계 학습 라이브러리와 소프트웨어 애플리케이션의 통합이 복잡하다.
* 엔터프라이즈 관련 데이터로 에이전트를 학습시키는 데 기술적 어려움이 있다.

### 4. 제한된 컴퓨팅 리소스

* 딥 러닝 AI 에이전트 학습 및 배포에 상당한 컴퓨팅 리소스가 필요하다.
* 온프레미스 구현 시 비싼 인프라 투자 및 유지 관리 부담이 있다.
* 쉽게 확장할 수 없는 제약사항이 존재한다.

## AI 에이전트의 단점

### 1. 불확정성

* 매 실행마다 다른 경로를 거쳐서 다른 결과를 출력한다.
* 일관성 있는 결과 보장이 어렵다.

### 2. 시간과 비용

* 한번의 사용자 목표 해결을 위해 여러 번의 LLM 호출이 필요하다.
* 상대적으로 높은 시간과 비용이 소모된다.

### 3. 사용할 수 있는 툴에 종속적

* AI 에이전트의 성능이 사용 가능한 도구들에 크게 의존한다.
* 도구의 품질과 범위가 에이전트 성능을 좌우한다.

## 결론

AI 에이전트는 **자율적으로 목표를 달성하는 지능형 소프트웨어 시스템**으로, 현대 비즈니스에서 생산성 향상과 고객 경험 개선에 핵심적인 역할을 하고 있다.

### 핵심 포인트

* **자율성과 지능성**을 바탕으로 복잡한 작업을 독립적으로 수행한다.
* **다양한 도구와 시스템을 연동**하여 종합적인 솔루션을 제공한다.
* **지속적인 학습과 개선**을 통해 성능을 향상시킨다.
* **기술적 도전과 한계**를 인식하고 적절한 대응 방안 마련이 필요하다.

AI 에이전트는 미래의 디지털 워크플로우에서 필수적인 요소가 될 것이며, 적절한 설계와 구현을 통해 조직의 디지털 트랜스포메이션을 가속화할 수 있다.



# LangGraph

https://docs.langchain.com/oss/python/langgraph/overview

### LangGraph 소개

* **LangGraph란?** LangGraph는 LangChain에서 제공하는 **멀티 에이전트 오케스트레이션 프레임워크**이다. 대형 언어 모델(LLM) 에이전트들을 **그래프 형태**로 연결하여 복잡한 작업을 수행하도록 설계되었다.
* **배경:** 하나의 LLM 에이전트로 해결하기 어려운 복잡한 문제를 여러 전문 에이전트로 분할해 협업시키는 아이디어에서 출발했다. LangGraph는 이러한 **다중 에이전트 시스템**을 체계적으로 구현하기 위한 도구이다.
* **주요 목적:** 대화형 에이전트, 복잡한 태스크 자동화, 사용자 정의 LLM 파이프라인 등 **AI 워크플로우를 확장 가능**하게 구축하는 토대를 제공한다.

### 철학과 배경 (Why LangGraph?)

* **멀티 에이전트 철학:** “혼자서는 어려운 일을 여럿이 분담하면 더 잘할 수 있다.” LangGraph는 **전문화된 에이전트들의 협업**을 통해 성능을 향상시키는 철학을 담고 있다. 한 에이전트가 모든 일을 다 하는 대신, 각 에이전트가 특정 역할에 집중하도록 분리한다.
* **상태 기계 관점:** LangGraph의 그래프 접근법은 **상태 기계(State Machine)** 개념과 유사하다. 각 에이전트 노드는 하나의 상태처럼 동작하고, 에이전트 간 전이는 상태 전이처럼 관리된다. 이러한 **구조적 설계**를 통해 복잡한 상호작용을 체계적으로 제어할 수 있다.
* **배경 및 등장:** Autogen, CrewAI 등 다양한 멀티 에이전트 프레임워크가 등장하는 가운데, LangChain 팀은 LangGraph를 통해 **그래프 기반 멀티 에이전트 워크플로우**의 이점을 강조했다. 이는 **유연한 흐름 제어**와 **공유 상태 관리**를 지원하여 기존 연속 실행(seq2seq) 방식의 한계를 보완한다.

### 그래프 구조 – 노드(Node)와 엣지(Edge)

* **노드 (Node):** 그래프의 노드는 각각 **독립된 에이전트**를 나타낸다. 각 에이전트는 자체 프롬프트, LLM 모델, 그리고 전용 **도구(tool)**들을 가질 수 있다. 예를 들어, 정보 검색을 담당하는 에이전트, 코드 실행을 담당하는 에이전트 등이 노드가 된다.
* **엣지 (Edge):** 노드 간 **이동 경로**를 나타낸다. 한 에이전트의 출력이 어느 다음 에이전트로 전달될지를 결정한다. 엣지는 **방향성**을 가지며, **조건부 제어**가 가능하다. 즉, 현재 상태나 메시지 내용에 따라 다른 경로를 선택할 수 있다 (분기/반복 구현 가능).
* **그래프 vs 순차 체인:** 전통적인 체인(chain)은 고정된 순서를 따르지만, **그래프 구조**에서는 유연한 경로 선택과 **분기/병렬 흐름**이 가능하다. 이를 통해 **복잡한 워크플로우**를 자연스럽게 표현할 수 있다.

### 상태 기반 에이전트 설계

* **공유 상태 (State):** LangGraph의 핵심은 에이전트들 사이에 공유되는 **그래프 상태**이다. 이 상태에는 대화 메시지, 중간 결과, 기타 컨텍스트 데이터가 포함된다. 모든 에이전트 노드는 이 **공용 메모리**를 통해 서로 소통한다.
* **에이전트 실행 모델:** 각 에이전트 노드는 자신의 역할에 맞는 **LLM 프롬프트**를 받아 상태를 처리하고, 새로운 정보를 상태에 추가한다. 예를 들어, ‘검색’ 에이전트는 질문을 받아 검색 결과를 상태에 추가하고, ‘답변’ 에이전트는 상태에 있는 검색 결과를 읽어 답을 생성한다.
* **상태 업데이트와 전이:** 에이전트가 작업을 마치면 상태(예: 메시지 목록, 작업 완료 플래그 등)를 갱신하고, **다음으로 어느 노드로 이동할지 결정**한다. LangGraph에서는 이 전이를 쉽게 정의할 수 있으며, 각 노드의 실행이 끝날 때 **어떤 상태이면 어느 노드로 이동**같은 로직을 구현한다. (예: 상태에 `sender="Researcher"`이면 다음은 `Chart Generator` 노드로 등)

### 메시지 흐름과 제어 흐름

* **메시지 전달:** LangGraph 그래프의 동작 흐름은 **메시지 패싱**으로 이해할 수 있다. 사용자의 입력 메시지가 그래프에 들어오면, **엔트리 노드(entry point)** 에이전트가 이를 받아 처리한다. 그 에이전트의 출력 메시지가 그래프 상태에 추가되고, 정의된 엣지를 따라 다음 에이전트로 전달된다.
* **제어 흐름:** 엣지에는 **조건부 로직**을 부여할 수 있다. 예를 들어, 특정 에이전트의 출력 내용에 따라 **다른 에이전트로 분기**하거나 **그래프 실행을 종료(END)**할 수 있다. 별도의 “라우터(router)” 에이전트를 두어 메시지 내용을 해석하고 다음 경로를 결정하기도 한다. 이를 통해 동적 의사결정이 가능한 워크플로우를 구현한다.
* **예시 – Router 패턴:** 하나의 라우터 노드가 전체 메시지 내용의 맥락을 보고 “계속 진행”, “도구 호출”, “종료” 등의 신호를 결정하면, 각 신호에 대응되는 다음 노드로 이동한다. 예컨대, **계속 진행**이면 다음 전문 에이전트로, **도구 호출**이면 외부 툴 실행 노드로, **종료**이면 END로 가는 식이다. (이와 같은 구조로 에이전트 간 협업을 **유연하게 제어**한다.)

### 실무 적용 고려사항

* **종료 조건 설정:** 멀티 에이전트 시스템은 잘못하면 무한 루프에 빠질 수 있다. **그래프 종료 조건(END)**을 명확히 정의해야 한다. 예를 들어, 에이전트가 “최종 답변”이라는 특별 신호를 출력하면 그래프를 종료시키는 규칙을 두어야 한다.
* **프롬프트 설계:** 각 에이전트에 **명확한 역할 지시**를 하는 프롬프트를 작성해야 한다. 에이전트들이 공유 상태를 활용하면서도 자신의 책임만 수행하도록 지침을 준다. 예를 들어, 한 에이전트는 “질문을 분석하여 검색어 생성”, 다른 에이전트는 “검색 결과로부터 답 생성” 식으로 구체적인 지시가 필요하다.
* **오버헤드와 비용:** 여러 에이전트가 순차/병렬로 LLM을 호출하므로 **API 호출 비용**과 **응답 지연**이 늘어날 수 있다. 실무 적용 시에는 필요한 최소한의 에이전트로 구성하고, 각 호출에 타임아웃 등의 제어를 넣어 **효율성**을 관리해야 한다.
* **디버깅과 모니터링:** 그래프 구조상 흐름이 복잡해질 수 있으므로, **중간 상태 로깅**이나 **시각화**가 중요하다. LangGraph로 실행되는 동안 어떤 경로를 거쳤는지, 각 노드의 입력/출력은 무엇이었는지 추적하는 것이 디버깅에 도움이 된다. (전문 도구: Langfuse 등 모니터링 솔루션 참고)
* **모델 한계 및 상호작용:** 각 에이전트는 LLM의 한계를 지닌다. 예를 들어, 정보를 검색했지만 잘못 이해하거나, 두 에이전트가 같은 정보에 대해 상충된 결과를 낼 수 있다. 이러한 경우를 대비해 **휴먼 인터벤션(human-in-the-loop)**이나 추가 검증 단계를 그래프에 포함시키는 것도 고려해야 한다.

### 활용 사례 및 마무리

* **QA 시스템:** 사용자의 질문을 받아 **검색 에이전트**가 웹에서 자료를 찾고, **답변 에이전트**가 결과를 요약하여 답변하는 시스템이다. 예를 들어 하나의 에이전트가 “서울 인구 검색”, 다음 에이전트가 “검색 결과로 답 생성” 형태로 협업한다.
* **코드 생성 협업:** **코드 작성 에이전트**와 **코드 검토 에이전트**가 번갈아 작동하여 고품질 코드를 산출한다. 작성자가 코드를 생성하면 검토자가 오류를 지적하고 수정 요청을 하며, 다시 작성자가 개선하는 **반복 루프**를 그래프를 통해 구현할 수 있다.
* **문서 요약 파이프라인:** 긴 문서를 **분할 요약 에이전트**가 부분 요약하고, **통합 요약 에이전트**가 이를 종합하여 최종 요약을 생성한다. 이처럼 요약 작업을 단계별 에이전트로 분리하면 매우 긴 입력도 효율적으로 처리 가능하다.
* **그 외 응용:** 대화형 비서에서 질의 유형에 따라 여러 전문 에이전트(예: 일정 관리, 이메일 작성)를 전환하며 처리하거나, 복잡한 문제를 해결하기 위해 **계획 수립 에이전트**가 하위 작업을 **전문 에이전트 팀**에 할당하는 등 다양하게 활용된다. LangGraph의 **그래프 접근법**은 이러한 확장적인 시나리오를 효과적으로 지원한다.


LangChain만으로도 “한 번 호출해서 답하는” RAG/챗봇은 충분히 만들 수 있다.
그러나 LangGraph가 따로 존재하는 이유는 LLM 앱이 커질수록 ‘흐름(상태, 분기, 반복, 복구)’을 안정적으로 관리해야만 하기 때문이다.

- 상태(State) 누적이 표준화됨
	- 대화/중간결과/도구결과를 State에 계속 쌓고, 노드가 그걸 읽고/추가하는 구조라서 파이프라인이 커져도 덜 꼬임.
- 분기/루프/조건 실행이 쉬움
	- “의도 분류 → (검색/계산/상담) 분기”, “답변이 부족하면 재검색 루프” 같은 제어흐름을 코드로 깔끔하게 표현.
- 멀티스텝 에이전트에 강함
	- 계획→도구호출→검증→최종답 같은 여러 단계를 노드로 쪼개 디버깅/테스트가 쉬움.
- 에러 복구/재시도 설계가 쉬움
	- 특정 노드 실패 시 재시도/대체 경로 등 운영 관점의 안정성을 넣기 좋음.
- 테스트/관찰(Observability) 포인트가 명확
	- 각 노드 입출력이 분리되니까 “어디서 망가졌는지” 추적이 쉬움(특히 LangSmith랑 붙일 때).

결론적으로
- LangChain = 구성요소(모델/프롬프트/리트리버/툴) 중심,
- LangGraph = 그 구성요소들을 ‘상태를 가진 워크플로우’로 운영하기 위한 프레임워크.
