## Pinecone Index 생성

```python
from pinecone import Pinecone, ServerlessSpec

pc = Pinecone() # pinecone api key 인증
print(pc.list_indexes().names()) # ['winemag-data-130k-v2', 'pinecone-first']


# 인덱스명 ir
if 'ir' not in pc.list_indexes().names():
    pc.create_index(
        name='ir',
        dimension=1536,
        metric='cosine',
        spec=ServerlessSpec(
            region='us-east-1',
            cloud='aws'
        )
    )
    print('ir인덱스가 생성되었습니다.')
else:
    print('ir인덱스가 이미 존재합니다.')

```

## VectorStore 연결

```python
from langchain_openai import OpenAIEmbeddings
from langchain_pinecone import PineconeVectorStore

embedding_model = OpenAIEmbeddings()
vector_store = PineconeVectorStore(
    index_name='ir',
    embedding=embedding_model
)
```

## Document Upsert

```python
import pandas as pd

df = pd.read_csv('documents.csv')
df
```

```python
# Document 변환
from langchain_core.documents import Document

docs_to_index = []
for idx, row in df.iterrows():
    doc_id = row['doc_id']
    content = row['content']
    doc = Document(page_content=content, metadata={'doc_id': doc_id})
    docs_to_index.append(doc)

print(f'{len(docs_to_index)}개 문서 준비완료')
```

```python
# 문서 업서트
vector_store.add_documents(docs_to_index)
```

```python
# 벡터 검색

query = '제주도 여행'

results = vector_store.similarity_search_with_score(query, k=5)
for rank, (doc, score) in enumerate(results, 1):
    print(f'{rank}: Score {score}')
    print(f'doc_id: {doc.metadata["doc_id"]}')
    print(f'content: {doc.page_content}')
    print()
```
